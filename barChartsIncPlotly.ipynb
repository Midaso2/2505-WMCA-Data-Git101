# Bar Charts

| Library | Interactivity | Style | Use Case |
|---------|---------------|-------|----------|
| Matplotlib | Static | Classic | Precise control, basic charts |
| Plotly | Interactive | Modern | Hover Effects, zoom, tooltips |
| Seaborn | Static | Stylish | Statistical plots built on Matplotlib |

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Simple sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': [10, 25, 36]
})

# Matplotlib
plt.bar(data['Category'], data['Value'], color='skyblue')
plt.title('Matplotlib Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# Seaborn
sns.barplot(x='Category', y='Value', data=data, palette='viridis')
plt.title('Seaborn Bar Chart Example')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plotly
fig = px.bar(data, x='Category', y='Value', title="Plotly Bar Chart Example")
fig.show()
```
